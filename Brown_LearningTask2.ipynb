# BY BARANGAY IN DASMARINAS

```python
# Import CSV file
# Scope: Select 'Barangay' in Status
db = pd.read_csv('dasma_concensus.csv')
pd.set_option('display.max_rows', None)
df = db[(db['Status'] == 'Barangay')]

# Get a sample from the dataset only
df = df.sample(len(df) // 2)
# Sort dataframe
df = df.sort_values(by='Name')
```

```python
"""========== SCATTER PLOT =========="""
# Create an index column for numerical x-axis
df['Index'] = range(len(df))

# Set the figure size of the plot
plt.figure(figsize=(8, 10))

# Create the Scatter Plot for Cases
sns.scatterplot(data=df, x='Cases', y='Index', hue='Cases',
                palette='winter', marker='o', s=75, label='Cases', legend=False)

# Create the Scatter Plot for Recoveries
sns.scatterplot(data=df, x='Recoveries', y='Index', hue='Recoveries',
                palette='summer', marker='s', s=75, label='Recoveries', legend=False)

# Create the Scatter Plot for Deaths
sns.scatterplot(data=df, x='Deaths', y='Index', color='red',  hue='Deaths',
                palette='autumn_r', marker='x', s=75, linewidth=2, label='Deaths', legend=False)

# Set labels and title
plt.ylabel('Barangay')
plt.xlabel('Number of Cases')
plt.title('COVID-19 Cases by Barangay in Dasmarinas')

# Add a legend that combines all plots
plt.legend(title='Legend')

# Set y-ticks to correspond to Barangay names
plt.yticks(df['Index'], df['Name'], fontsize=8)
```

```python
"""========== LINE PLOT =========="""
# Create an index column for numerical x-axis
df['Index'] = range(len(df))

# Set the figure size of the plot
fig, ax = plt.subplots(figsize=(8,10))

# Create the Line Plot
sns.lineplot(data=df, x=df['Name'], y=df['Cases'], color='#003366', label='Cases')
sns.lineplot(data=df, x=df['Name'], y=df['Recoveries'], color='#003300', label='Recoveries')
sns.lineplot(data=df, x=df['Name'], y=df['Deaths'], color='#660000', label='Deaths')

# Set labels and title
plt.xlabel('Barangay')
plt.ylabel('Number of Cases')
plt.title('COVID-19 Cases by Barangay in Dasmarinas')

# Add a legend
plt.legend()

# Set x-ticks to correspond to Barangay names
plt.xticks(df['Index'], df['Name'], fontsize=8, rotation=90)
```

```python
"""========== BAR GRAPH =========="""
# Create an index column for numerical x-axis
df['Index'] = range(len(df))

# Create the bar plot using Seaborn
plt.figure(figsize=(8, 10))
sns.barplot(data=df, x='Cases', y='Index', hue='Cases', orient='h', palette='dark:#008080')
sns.barplot(data=df, x='Recoveries', y='Index', hue='Recoveries', orient='h', palette='dark:#FF7F50')
sns.barplot(data=df, x='Deaths', y='Index', hue='Deaths', orient='h', palette='dark:#DC143C')

# Set labels and title
plt.ylabel('Barangay')
plt.xlabel('Number of Cases')
plt.title('COVID-19 Cases by Barangay in Dasmarinas')

# Set y-ticks to correspond to Barangay names
plt.yticks(df['Index'], df['Name'], fontsize=8)

# Manually create the legend
legend_labels = ['Cases', 'Recoveries', 'Deaths']
colors = ['#008080', '#FF7F50', '#DC143C']

# Create custom legend handles
handles = [plt.Line2D([0], [0], color=color, lw=4) for color in colors]

# Add the custom legend
plt.legend(handles=handles, labels=legend_labels, title='Cases')
```

```python
"""========== PIE CHART =========="""
plt.figure(figsize=(8, 8))

colors = ['#003366','#004080','#0059b3','#0073e6']

# Create the Pie Chart
wedges, texts, autotexts = plt.pie(df['Cases'], labels=df['Name'], autopct='%1.1f%%', pctdistance = 0.75, startangle=140, colors=colors, label='Cases')

# Set font size for labels
for text in texts:
    text.set_fontsize(8)

# Set font size for percentages
for autotext in autotexts:
    autotext.set_fontsize(8)
    autotext.set_color('white')
    autotext.set_fontweight('bold')

# Set the title
plt.title('COVID-19 Case Distribution by Barangay in Dasmarinas')
```

```python
"""========== PIE CHART =========="""
plt.figure(figsize=(8, 8))

colors = ['#003300','#004d00','#006600','#009900']

# Create the Pie Chart
wedges, texts, autotexts = plt.pie(df['Recoveries'], labels=df['Name'], autopct='%1.1f%%', pctdistance = 0.75, startangle=140, colors=colors, label='Recoveries')

# Set font size for labels
for text in texts:
    text.set_fontsize(8)

# Set font size for percentages
for autotext in autotexts:
    autotext.set_fontsize(8)
    autotext.set_color('white')
    autotext.set_fontweight('bold')

# Set the title
plt.title('COVID-19 Recovery Distribution by Barangay in Dasmarinas')
```

```python
"""========== PIE CHART =========="""
plt.figure(figsize=(8, 8))

colors = ['#660000','#990000','#cc0000','#ff0000']

# Create the Pie Chart
wedges, texts, autotexts = plt.pie(df['Deaths'], labels=df['Name'], autopct='%1.1f%%', pctdistance = 0.75, startangle=140, colors=colors, label='Deaths')

# Set font size for labels
for text in texts:
    text.set_fontsize(8)

# Set font size for percentages
for autotext in autotexts:
    autotext.set_fontsize(8)
    autotext.set_color('white')
    autotext.set_fontweight('bold')

# Set the title
plt.title('COVID-19 Death Distribution by Barangay in Dasmarinas')
```

```python
"""========== HISTOGRAM PLOT =========="""
# Set up the figure and axis
plt.figure(figsize=(8, 6))

# Create the Histogram
sns.histplot(x='Cases', hue='Cases', data=df, kde=True, bins=20, multiple='stack', edgecolor='.3', linewidth=.5, palette='winter_r')

# Set labels and title
plt.ylabel('Count')
plt.xlabel('Number of Cases')
plt.title('Frequency of COVID-19 Cases in Dasmarinas')
plt.legend(['Cases'], title='Legend', loc='upper right')
```

```python
"""========== HISTOGRAM PLOT =========="""
# Set up the figure and axis
plt.figure(figsize=(8, 6))

# Create the Histogram
sns.histplot(x='Recoveries', hue='Recoveries', data=df, kde=True, bins=20, multiple='stack', edgecolor='.3', linewidth=.5, palette='summer_r')

# Set labels and title
plt.ylabel('Count')
plt.xlabel('Number of Recoveries')
plt.title('Recovery Patients from COVID-19 in Dasmarinas')
plt.legend(['Recoveries'], title='Legend', loc='upper right')
```

```python
"""========== HISTOGRAM PLOT =========="""
# Set up the figure and axis
plt.figure(figsize=(8, 6))

# Create the Histogram
sns.histplot(x='Deaths', hue='Deaths', data=df, kde=True, bins=20, multiple='stack', edgecolor='.3', linewidth=.5, palette='autumn_r', label='Recoveries')

# Set labels and title
plt.ylabel('Count')
plt.xlabel('Number of Deaths')
plt.title('Deaths due to COVID-19 in Dasmarinas')
plt.legend(['Deaths'], title='Legend', loc='upper right')
```
